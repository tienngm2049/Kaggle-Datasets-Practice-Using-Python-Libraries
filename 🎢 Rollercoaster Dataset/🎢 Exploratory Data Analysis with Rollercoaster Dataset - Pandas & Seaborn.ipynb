### 2.3 Handle Missing Values

There are several ways to handle missing values in a dataset. The choice of method depends on the nature of the data and the specific analysis or modeling task. Here are some common approaches:

1. **Dropping missing values**: If the missing values are relatively few and randomly distributed, you can simply drop the rows or columns containing missing values. This can be done using the `dropna()` function in pandas.

   ```python
   df.dropna()  # Drop rows with any missing value
   df.dropna(axis=1)  # Drop columns with any missing value
   ```

   However, be cautious when using this method, as dropping too many observations can lead to loss of valuable information.

2. **Imputation**: Missing values can be replaced with estimated values. Some common imputation techniques include:

   - **Mean or median imputation**: Replace missing values with the mean or median of the available data for that variable.

     ```python
     df['column_name'].fillna(df['column_name'].mean(), inplace=True)
     ```

   - **Mode imputation**: Replace missing values with the mode (most frequent value) of the variable.

     ```python
     df['column_name'].fillna(df['column_name'].mode()[0], inplace=True)
     ```

   - **Forward or backward fill**: Fill missing values with the last observed value (forward fill) or the next observed value (backward fill).

     ```python
     df.fillna(method='ffill', inplace=True)  # Forward fill
     df.fillna(method='bfill', inplace=True)  # Backward fill
     ```

   - **Interpolation**: Fill missing values by interpolating between existing values. This can be useful for time series data or data with a specific order.

     ```python
     df.interpolate(inplace=True)
     ```

   Imputation methods should be used with caution, as they introduce assumptions and may impact the distribution and relationships in the data.

3. **Indicator variables**: Create new binary variables that indicate whether a value was missing or not. This approach allows the missingness to be included as a feature in the analysis.

   ```python
   df['column_name_isnull'] = df['column_name'].isnull().astype(int)
   ```

4. **Advanced techniques**: There are more advanced techniques available for handling missing data, such as multiple imputation or using machine learning algorithms to predict missing values based on other variables. These methods are more complex and may require additional libraries or expertise.

Remember to carefully consider the nature of the missing data and the implications of each handling method for your specific analysis or modeling task.

````python
df.columns
````

````python
df = df.loc[~df.duplicated(subset = ['Coaster_Name','Location', 'Opening_Date'])].reset_index(drop=True).copy()
````

The code `df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True)` performs the following operations:

1. `df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])` checks for duplicate rows in the DataFrame `df` based on the specified subset of columns, which in this case are 'Coaster_Name', 'Location', and 'Opening_Date'. The `duplicated()` function returns a boolean Series where True represents duplicate rows and False represents non-duplicate rows.

2. `~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])` uses the negation operator (~) to invert the boolean values in the Series. This will give us a Series where True represents non-duplicate rows and False represents duplicate rows. The tilde (~) operator is used to negate the values in the Series.

3. `df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])]` selects the rows from the DataFrame `df` where the corresponding values in the Series are True. In other words, it filters out the duplicate rows and keeps only the non-duplicate rows.

4. `reset_index(drop=True)` resets the index of the resulting DataFrame. The `drop=True` parameter ensures that the old index is not added as a new column in the DataFrame.

Overall, this code snippet removes duplicate rows from the DataFrame based on the specified subset of columns ('Coaster_Name', 'Location', 'Opening_Date'), and returns a new DataFrame with only the non-duplicate rows.

````python
# Get correlation data
corr = df[['Year_Introduced','Speed_Mph','Height_Ft','Inversions','Gforce']].dropna().corr()
corr_matrix = corr.iloc[1:,:-1]
corr_matrix
````

````python
# ones_like can build a matrix of booleans (True,False) with the same shape as data
ones = np.ones_like(corr, dtype=bool)
ones
````

````python
# np's triu: return only triangle matrix
mask = np.triu(ones)
adjust_mask = mask[1:,:-1]
adjust_mask
````

````python
sns.heatmap(corr_matrix, cmap = 'Blues', annot= True, mask = adjust_mask)
````

## Step 5: Ask a Question about the dataÂ¶
Try to answer a question you have about the data using a plot or statistic.

**"What are the locations with the fastest roller coasters (minimum of 10)?"**

````python
groupby = df.query("Location != 'Other'") \
            .groupby('Location')['Speed_Mph'] \
            .agg(['mean','count']) \
            .query('count >= 10') \
            .sort_values('mean', ascending=False)
groupby
````

````python
sns.barplot(x='mean', y=groupby.index, data=groupby, color = "brown")
plt.xlabel('Mean Speed (mph)')
plt.ylabel('Location')
plt.title('Mean Speed of Coasters by Location');
````
